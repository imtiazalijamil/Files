```python
from lxml import html
import requests
from time import sleep
records=1
data = []
with open("url.txt") as in_file:
    for url in in_file:
        page = requests.get(url.strip())
        refLink=url.strip()
        content = html.fromstring(page.content)
        names = content.xpath('/html/body/div[6]/div/div[2]/div[2]/h2/text()')
        urls= content.xpath("/html/body/div[6]/div/div[2]/div[2]/a/@href")
        aURL = "https://superexpo.com/" + urls[0]
        data.append((names[0],aURL,refLink))
        print(records,names[0],aURL)
        records=records+1
        #sleep(2)
import pandas as pd 
df = pd.DataFrame(data)
df.to_csv("7583A7583.csv")
```
