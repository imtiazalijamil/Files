```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from time import sleep
import pandas as pd

op = webdriver.ChromeOptions()
op.add_argument =("--headless")
op.add_argument =("--no-sandbox")

# path_to_chromedriver = 'C:\chromedriver.exe' # change path as needed
# browser = webdriver.Chrome(executable_path = path_to_chromedriver,chrome_options=op)

browser = webdriver.Chrome(executable_path=r'C:\chromedriver.exe', options=op)
baseurl = 'https://superexpo.com/en/login?r=https://superexpo.com/en/exhibitor/107831/priminer-machine-tools/'
browser.get(baseurl)
start_time = time.time()
#browser.find_element_by_name("login").send_keys("mebas18955@nowdigit.com")
browser.find_element_by_name("pass").send_keys("mebas18955@nowdigit.com")
browser.find_element_by_name("submit").click()
sleep(5)

data = []
records=0
with open("urls.txt") as in_file:
    
    for url in in_file:
        
        browser.get(url.strip())
        refLink=url.strip()
        cEmail = browser.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[1]/div[4]").text
    
        if cEmail=="for logged user only":
            print("website is not logged in")
            browser.get(baseurl)
            start_time = time.time()
            browser.find_element_by_name("login").send_keys("mebas18955@nowdigit.com")
            browser.find_element_by_name("pass").send_keys("mebas18955@nowdigit.com")
            browser.find_element_by_name("submit").click()
            sleep(5)
        else:
        
            refLink=url.strip()
            #c1 = browser.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[1]/div[4]").text
            cName = browser.find_element_by_class_name("contact-name").text
            cAddress = browser.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[1]/div[1]").text
            cPhone = browser.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[1]/div[3]").text
            cFax = browser.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[1]/div[2]").text
            cEmail = browser.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[1]/div[4]").text
            cWeb = browser.find_element_by_xpath("/html/body/div[6]/div/div[2]/div[1]/div[5]").text
            data.append((cName,cAddress,cPhone,cFax,cEmail,cWeb,refLink))
            sleep(2)
            elapsed_time = time.time() - start_time
            records=records+1
            print("Total Rows Processed: ",records,"Time Spent: ",time.strftime("%H:%M:%S", time.gmtime(elapsed_time)))

df = pd.DataFrame(data)
df.to_csv("513Data.csv")
browser.quit()
            
```
